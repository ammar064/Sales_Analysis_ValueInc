```python
# get the Revenue ,Cost ,and Profit
data['Revenue']=data['NumberOfItemsPurchased']*data['SellingPricePerItem']
data['TransactionCost']=data['NumberOfItemsPurchased']*data['CostPerItem']
data['Profit']=data['Revenue']-data['TransactionCost']
data.head()
```

```python
# we need to calculate markup which eqaul to  (Revenue-cost) /cost
data['Markup']=(data['Revenue']-data['TransactionCost'])/data['TransactionCost']
```

```python
data.head()
```

```python
#using round function in Markup column
data['Markup']=round(data['Markup'],2)
#show frist 5 rows
data.head()
```

```python

#handling with date and chane the type of data to string
data['Day'].dtype
data['Day']=data['Day'].astype(str)
```

```python
#the same for year
data['Year'].dtype
data['Year']=data['Year'].astype(str)
```

```python
#combing data field 
data['my_date']=data['Day']+'-'+data['Month']+'-'+data['Year']
#show frist 5 rows
data.head()
```

```python
#clean ClientKeywords Column
# use split function
data['AgeOfClient']=data['ClientKeywords'].str.split(',',expand=True)[0]
data['CleintType']=data['ClientKeywords'].str.split(',',expand=True)[1]
data['ContractLenght']=data['ClientKeywords'].str.split(',',expand=True)[2]
data.head(1)
```

```python
data['ContractLenght']= data['ContractLenght'].str.replace(']','')
data['ContractLenght']=data['ContractLenght'].str.replace("'",'')
data['AgeOfClient']=data['AgeOfClient'].str.replace("'",'')
data['AgeOfClient']=data['AgeOfClient'].str.replace("[",'')
data['CleintType']=data['CleintType'].str.replace("'",'')
data.head(1)
```

```python
#using the lowerCase Function

data['ItemDescription']=data['ItemDescription'].str.lower()
```

```python
# How to merge anther files
#bring new dataset called season
season= pd.read_csv('/kaggle/input/seasons/value_inc_seasons.csv', sep =';',encoding ='utf-8')
season
```

```python
#merge file 
new_data=pd.merge(data,season,on='Month')
new_data.head()
```

```python
#drop some columns
new_data=new_data.drop(columns=['Year','Month','Day','ClientKeywords'],axis=1)
new_data.head()
```

```python
#Export into CSV 
new_data.to_csv('ValueInc_cleaned.csv',index=False)
```
